```python
df["Installs"].unique()
```

```python
df["Installs"].value_counts()
```

```python
df["Installs"]=df["Installs"].str.replace("+","")
df["Installs"]=df["Installs"].str.replace(",","")
```

```python
df["Installs"].unique()
```

```python
df["Installs"]=df["Installs"].astype(int)
```

```python
df.info()
```

```python
# category with maximum number of installations
```

```python
df1=df.groupby(["Category"])["Installs"].sum().sort_values(ascending=False)
```

```python
print(f"top 10 category with maximum number of installs : \n\n{df1.head(10)}")
```

```python
sns.set_context("talk")
sns.set_style("darkgrid")

sns.barplot(x="Category",y="Installs",data=df1.head(10))
plt.xticks(rotation=90)
plt.title("Top 10 Category with maximum installs")
plt.show()
```

```python
sns.set_context("talk")
sns.set_style("darkgrid")

sns.barplot(y="Category",x="Installs",data=df1.head(10))
plt.xticks(rotation=90)
plt.title("Top 10 Category with maximum installs")
plt.show()
```
