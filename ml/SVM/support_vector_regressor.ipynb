```python
x=df.drop("total_bill",axis=True)
y=df["total_bill"]
```

```python
x.head()
```

```python
y.head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=43)
```

```python
df.columns
```

```python
le1=LabelEncoder()
le2=LabelEncoder()
le3=LabelEncoder()
```

```python
x_train["sex"]=le1.fit_transform(x_train["sex"])
x_train["smoker"]=le2.fit_transform(x_train["smoker"])
x_train["time"]=le3.fit_transform(x_train["time"])
```

```python
x_test["sex"]=le1.transform(x_test["sex"])
x_test["smoker"]=le2.transform(x_test["smoker"])
x_test["time"]=le3.transform(x_test["time"])
```

```python
encoder1=OneHotEncoder(drop="first",sparse_output=False)
```

```python
# Fit and transform the 'day' column
day_train_encoded = encoder1.fit_transform(x_train[["day"]])
day_test_encoded = encoder1.transform(x_test[["day"]])

# Convert encoded output to DataFrame
day_train_df = pd.DataFrame(day_train_encoded,
                            columns=encoder1.get_feature_names_out(["day"]),
                            index=x_train.index)

day_test_df = pd.DataFrame(day_test_encoded,
                           columns=encoder1.get_feature_names_out(["day"]),
                           index=x_test.index)

# Drop original 'day' column
x_train = x_train.drop(columns="day")
x_test = x_test.drop(columns="day")

# Concatenate encoded columns
x_train = pd.concat([x_train, day_train_df], axis=1)
x_test = pd.concat([x_test, day_test_df], axis=1)
```

```python
x_train.head()
```

```python
model=SVR()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
r2=r2_score(y_test,y_pred)
mse=mean_squared_error(y_test,y_pred)
mae=mean_absolute_error(y_test,y_pred)
```
