```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score,roc_curve
```

```python
x,y=make_classification(
    n_samples=1000,
    n_features=10,
    n_classes=2,
    random_state=43
)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=43)
```

```python
regressor=LogisticRegression()
regressor.fit(x_train,y_train)
```

```python
dummy_prob=[0 for i in range(len(y_test))]
```

```python
len(dummy_prob),y_test.shape
```

```python
dummy_prob
```

```python
model_prob=regressor.predict_proba(x_test)
```

```python
model_prob
```

```python
model_prob.shape
```

```python
model_prob=model_prob[:,1]
```

```python
model_prob
```

```python
dummy_fpr,dummy_tpr,threshold1=roc_curve(y_test,dummy_prob)
model_fpr,model_tpr,threshold2=roc_curve(y_test,model_prob)
```

```python
plt.plot(dummy_fpr,dummy_tpr,linestyle="-",label="dummy model")
plt.plot(model_fpr,model_tpr,marker=".",label="logistic model")
plt.xlabel("false positive rate")
plt.ylabel("true positive rate")
plt.legend()
plt.show()
```

```python
fig=plt.figure(figsize=(20,20))
ax=fig.add_subplot(111)

for fpr, tpr, thresh in zip(model_fpr, model_tpr, threshold2):
    ax.annotate(f"{np.round(thresh, 2)}", xy=(fpr, tpr), fontsize=8)


plt.plot(dummy_fpr,dummy_tpr,linestyle="-",label="dummy model",color="green")
plt.plot(model_fpr,model_tpr,marker=".",label="logistic model",color="yellow")

plt.xlabel("false positive rate")
plt.ylabel("true positive rate")
plt.legend()
plt.show()
```
