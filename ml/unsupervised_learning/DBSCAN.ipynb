```python
x,y=make_moons(
    n_samples=300,
    shuffle=True,
    noise=0.05,
    random_state=42
)
```

```python
x.shape
```

```python
scaler=StandardScaler()
x_scaled=scaler.fit_transform(x)
```

```python
pca=PCA(n_components=2)
x_pca=pca.fit_transform(x_scaled)
```

```python
cluster=DBSCAN(
    eps=0.9,
    min_samples=5,
    metric='euclidean',
    n_jobs=-1
)
```

```python
cluster.fit(x_pca)
```

```python
plt.scatter(x_pca[:,0],x_pca[:,1],c=cluster.labels_)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.show()
```
